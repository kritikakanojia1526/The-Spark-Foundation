Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
import pandas as pd
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
# checking rows and column
df.shape
```

```python
df.describe()
```

```python
x=df.drop(["Species"],axis=1)
x=x.drop(["Id"],axis=1)
y=df[["Species"]]
```

```python
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
x_names=['SepalLength','SepalWidth','PetalLength','PetalWidth']
y_names=['Iris-Setosa','Iris-Versicolor','Iris-Virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(model, feature_names = x_names, class_names = y_names, filled = True);
```

```python
#Example
print(model.predict([[1.2,2.4,4.5,8.9]]))
```
